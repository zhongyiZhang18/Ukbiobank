```python
import pandas as pd
gd_loss = pd.read_csv('g&d_loss.csv') # 读取训练数据
print(gd_loss.shape)
```

```python
l = gd_loss.shape[0]
x_axix = np.arange(1,l+1)
x_axix
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

l = gd_loss.shape[0]
x_axix = np.arange(1,l+1)
plt.title('GAN LOSS ')
plt.plot(x_axix, gd_loss['g_loss'], color='green', label='gloss')
plt.plot(x_axix, gd_loss['d_loss'], color='red', label='dloss')

plt.legend() # 显示图例

plt.xlabel('iteration times(/10)')
plt.ylabel('loss of generator and discriminator')
plt.show()
```
